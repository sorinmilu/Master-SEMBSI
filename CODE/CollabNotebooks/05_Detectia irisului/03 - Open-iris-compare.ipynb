```python
import cv2
import matplotlib.pyplot as plt
from google.colab import files
from IPython.display import display
import numpy as np
from google.colab.patches import cv2_imshow
import iris
```

```python
print("Incarcati prima imagine de test")
uploaded_test_image_1 = files.upload()  # Allows user to upload a file
# Get the first uploaded file
uploaded_test_image_1_path = list(uploaded_test_image_1.keys())[0]
img_pixels_1 = cv2.imread(uploaded_test_image_1_path, cv2.IMREAD_GRAYSCALE)
```

```python
print("Incarcati a doua imagine de test")
uploaded_test_image_2 = files.upload()  # Allows user to upload a file
# Get the first uploaded file
uploaded_test_image_2_path = list(uploaded_test_image_2.keys())[0]
img_pixels_2 = cv2.imread(uploaded_test_image_2_path, cv2.IMREAD_GRAYSCALE)
```

```python
combined_image = np.hstack((img_pixels_1, img_pixels_2))

# Display the combined image
cv2_imshow(combined_image)
```

```python
iris_pipeline = iris.IRISPipeline()

output_1 = iris_pipeline(img_pixels_1, eye_side="left")
output_2 = iris_pipeline(img_pixels_2, eye_side="left")
```

```python
matcher = iris.HammingDistanceMatcher()
```

```python
distance = matcher.run(output_1['iris_template'], output_2['iris_template'])

print(f"Distanta dintre cele doua imagini {distance:.4f}")
if distance < 0.38:
  print("acelasi ochi")
else:
  print("ochi diferiti")
```
