```python
import cv2
import matplotlib.pyplot as plt
from mtcnn import MTCNN
from google.colab import files
from IPython.display import display
from google.colab.patches import cv2_imshow
```

```python
detector = MTCNN()
```

```python
print("Incarcati o imagine de test")
uploaded_test_image = files.upload()  # Allows user to upload a file
# Get the first uploaded file
uploaded_test_image_path = list(uploaded_test_image.keys())[0]
```

```python
img = cv2.imread(uploaded_test_image_path)
cv2_imshow(img)
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
faces = detector.detect_faces(img_rgb)
```

```python
image_copy = img.copy()
for face in faces:
    # Get bounding box coordinates
    x, y, width, height = face['box']

    # Draw the bounding box
    cv2.rectangle(image_copy, (x, y), (x + width, y + height), (0, 255, 0), 2)

    # Draw landmarks (5 points)
    for landmark in face['keypoints'].values():
        cv2.circle(image_copy, landmark, 4, (0, 0, 255), -1)  # Red dots for landmarks
```

```python
cv2_imshow(image_copy)
```
