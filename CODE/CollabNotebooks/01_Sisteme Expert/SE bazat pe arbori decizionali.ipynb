```python
plants = {
    "question": "Does it have flowers?",
    "yes": {
        "question": "Is it a tree?",
        "yes": {
            "question": "Does it produce fruit?",
            "yes": "apple tree",
            "no": "oak tree"
        },
        "no": {
            "question": "Is it a small plant?",
            "yes": {
                "question": "Does it have a strong fragrance?",
                "yes": "lavender",
                "no": "sunflower"
            },
            "no": {
                "question": "Is it a climbing plant?",
                "yes": "ivy",
                "no": "rose"
            }
        }
    },
    "no": {
        "question": "Does it have leaves?",
        "yes": {
            "question": "Is it a fern?",
            "yes": "fern",
            "no": {
                "question": "Does it have a woody stem?",
                "yes": "cactus",
                "no": "grass"
            }
        },
        "no": {
            "question": "Does it have a thick stem?",
            "yes": "cactus",
            "no": "moss"
        }
    }
}
```

Funcție utilitară care pune întrebările si returneaza raspunsurile

```python
def ask(question):
    while True:
        answer = input(question + " (yes/no): ").strip().lower()
        if answer in ('yes', 'y'):
            return True
        elif answer in ('no', 'n'):
            return False
        else:
            print("Please answer yes or no.")
```

Funcția principală care traversează întreg arborele și pune întrebările. Aceasta poate fi folosită cu orice arbore decizioan pe bază de da/nu

```python
def traverse(tree):
    # If this is a leaf node (just an animal string), guess
    if isinstance(tree, str):
        print(f"Is it a {tree}?")
        return

    # Pune întrebarea curentă
    answer = ask(tree['question'])

    # Ne mutăm pe următorul nod
    next_node = tree['yes'] if answer else tree['no']
    traverse(next_node)
```

```python
print("Raspundeti la intrebari!\n")
traverse(plants)
```
